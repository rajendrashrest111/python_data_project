```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
median_salary_year= df['salary_year_avg'].median()
```

```python
median_salary_hour= df['salary_hour_avg'].median()
```

# in here we are going to use df.dropna(): drop missing values. we are handling missing data

df_cleaned= df.dropna(how='all')
# then we are filling the NA values using fillna()
# then we are removing the duplicate items like advertisment for the same job from the same company

```python
df_cleaned= df.dropna(how='all')
```

```python
df_cleaned

```

```python
df_filled= df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled.loc[:19,'salary_year_avg': 'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates() # without using subsets

print('length of orginal df: ', len(df_filled))
print('length of drop duplicate_df: ', len(df_unique))
print('rows dropped:                ', len(df_filled)-len(df_unique))
```

```python
#now we are using the subsets
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])
print('length of orginal df: ', len(df_filled))
print('length of drop duplicate_df: ', len(df_unique))
print('rows dropped:                ', len(df_filled)-len(df_unique))
```
