```python
# here, we are doing data sample and copying method
# .sample() is a method in pandas used to randomly select rows or columns from a dataframe or series
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(10)
```

```python
df.sample(5, random_state=42)
# random state ensures the same sample everytime running the code. the number on it can be any number but when used multiple times better to use the same

```

```python
df.sample(frac=0.01) # ~ 1% of the rows
```

```python
df.sample(3, axis=1)   # randomly selects 3 columns
```

```python
# In pandas, .copy() creates a new, independent copy of a DataFrame (or Series).
# That means the new object does not share memory with the original one.
# Without .copy(), pandas often just creates a view — a "window" into the same data. If you change one, you might accidentally change the other.
```

```python
df_subset = df[['job_title', 'salary_year_avg']].copy()
df_subset['salary_year_avg'] = df_subset['salary_year_avg'].fillna(0)
```
